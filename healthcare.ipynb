```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("healthcare_dataset.csv")
```

```python
# Explore correlation between variables
correlation = data.groupby(['Blood Group Type', 'Gender'])['Medical Condition'].value_counts().unstack()
correlation.plot(kind='bar', stacked=True)
plt.xlabel('Blood Group Type, Gender')
plt.ylabel('Count of Medical Conditions')
plt.title('Correlation between Blood Group Type, Gender, and Medical Condition')
plt.show()
```

```python
contingency_table = pd.crosstab(data["Blood Group Type"], [data["Gender"], data["Medical Condition"]])
contingency_table.head()  # View the first few rows
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table.iloc[:-1, :-1], annot=True, fmt="d")  # Exclude totals
plt.title("Co-relation Between Blood Group, Gender, and Medical Condition")
plt.show()
```

```python
average_billing = data.groupby("Medical Condition")["Billing Amount"].mean().reset_index()
average_billing.sort_values(by="Billing Amount", inplace=True, ascending=False)  # Sort by highest average
average_billing.head()  # View the top few rows
```
